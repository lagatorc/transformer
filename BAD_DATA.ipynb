```python
def read_files(paths):
    df_list = []
    for path in paths:
        print(path)
        temp = pd.read_csv(path, error_bad_lines=False)
        temp = temp.rename(columns={'Semiotic Class':'semiotic_token', 'Input Token':'input_token', 'Output Token':'output_token'})
        df_list.append(temp)
    return df_list
```

```python
train_list = read_files(train_files)
```

```python
test_list = read_files(test_files)
```

```python
val_list = read_files(val_files)
```

```python
def create_sentences(df)    
    semiotic_sentence = ''
    input_sentence = ''
    output_sentence = ''

    s_sent = []
    i_sent = []
    o_sent = []

    for row in df.iterrows():
        if row[1].semiotic_token == '<eos>':
            s_sent.append(semiotic_sentence[:-1])
            semiotic_sentence = ''
            i_sent.append(input_sentence[:-1])
            input_sentence = ''
            o_sent.append(output_sentence[:-1])
            output_sentence = ''
            continue

        semiotic_token = str(row[1].semiotic_token)
        input_token = str(row[1].input_token)
        output_token = str(row[1].output_token)
        if row[1].output_token == '<self>':
            output_token = input_token
        semiotic_sentence = semiotic_sentence + semiotic_token + ' '
        input_sentence = input_sentence + input_token + ' '
        output_sentence = output_sentence + output_token + ' '
```
